### Bài 1. (3 điểm)

Một công ty nhận sản phẩm từ ba xưởng $A,B,C$. Tỉ lệ sản lượng của ba xưởng lần lượt là $40\%$, $35\%$, $25\%$. Tỉ lệ sản phẩm lỗi của ba xưởng lần lượt là $2\%$, $3\%$, $5\%$.

a) Chọn ngẫu nhiên một sản phẩm của công ty. Tính xác suất để sản phẩm đó bị lỗi.

### Bài 2. (3 điểm)

Một hộp có $8$ viên bi gồm hai màu đỏ và xanh, nhưng chưa biết chính xác số bi đỏ. Gọi $A_k$ là biến cố “trong hộp có đúng $k$ viên bi đỏ”, với $k=0,1,\ldots,8$. Giả sử tiên nghiệm của các biến cố $A_k$ là như nhau.

$$
R,\ R,\ B.
$$

Người ta bốc ngẫu nhiên lần lượt $3$ viên bi **không hoàn lại** và quan sát được dãy màu như trên.

a) Hãy viết công thức hậu nghiệm $P(A_k\mid R,R,B)$ theo $k$.

b) Giá trị nào của $k$ có hậu nghiệm lớn nhất?

c) Vẽ biểu đồ cột biểu diễn phân phối hậu nghiệm.


### Bài 3. (2 điểm)

Trong một **thí nghiệm y học**, các bệnh nhân được đưa vào thử nghiệm tuần tự cho đến khi xuất hiện bệnh nhân đầu tiên có phản ứng tốt với thuốc. Gọi $p$ là xác suất một bệnh nhân bất kỳ đáp ứng tốt với thuốc. Trước khi quan sát dữ liệu, giả sử:

$$
p \sim \mathrm{Beta}(3,2).
$$

Gọi $Y$ là số bệnh nhân cần theo dõi cho đến khi xuất hiện **ca đáp ứng tốt đầu tiên**, và giả sử

$$
Y\mid p \sim \mathrm{Geometric}(p).
$$

Quan sát được rằng phải đến **bệnh nhân thứ 4** mới xuất hiện ca đáp ứng tốt đầu tiên.


a) Hãy xác định phân phối hậu nghiệm của $p$.


b) Tính kỳ vọng hậu nghiệm của $p$ và vẽ trên cùng một hệ trục mật độ của phân phối tiên nghiệm và hậu nghiệm.


### Bài 4. (2 điểm)

Gọi $a,b$ lần lượt là chữ số gần cuối, chữ số cuối trong MSSV của anh chị. Chẳng hạn MSSV là $20123671$ thì $a=7,b=1$.

Trong một bối cảnh **kinh tế**, giả sử số doanh nghiệp nhỏ **giải thể mới** trong mỗi tháng tại một quận được mô hình bằng phân phối Poisson với tham số $\lambda$. Trước khi quan sát dữ liệu mới, giả sử:

$$
\lambda \sim \mathrm{Gamma}(\text{có kỳ vọng } 3,\ \text{độ lệch chuẩn } 1).
$$

Trong $3$ tháng gần đây, số doanh nghiệp nhỏ giải thể mới lần lượt là $a,b,2$.


a) Hãy xác định phân phối hậu nghiệm của $\lambda$.


b) Tính kỳ vọng hậu nghiệm của $\lambda$ và vẽ trên cùng một hệ trục mật độ của phân phối tiên nghiệm và hậu nghiệm.
