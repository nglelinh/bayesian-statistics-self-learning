**Bài 7.** (Cặp biến cố độc lập dựa trên một biến cố khác) 

Ta biết rằng nếu $A,B$ độc lập thì $P(A.B)=P(A)P(B)$ và $P(A|B)=P(A)$. Từ đó với biến cố $C$ có $P(C)>0$ thì $$P(A.B|C) = P(A|C)P(B|C).$$
Từ đó, ta có thể chứng minh được $$P(A|B,C)=P(A|C).$$
