### Bài 3. (2 điểm)

Xét một phân phối tiên nghiệm $X \sim N(\mu = 16, \sigma^2 = 0.5).$ Giả sử một mẫu ngẫu nhiên gồm $15$ quan sát bao gồm các giá trị có phân phối chuẩn là 
$$16.09, 15.99, 15.77, 16.04, 16.18, 15.72, 15.98, 16.39, 15.51, 16.22, 15.96, 15.81, 15.86, 16.46, 15.85.$$

a) Dựa vào công thức liên hợp đã biết của phân phối chuẩn, hãy xác định phân phối hậu nghiệm. Gợi ý: nếu có phân phối tiên nghiệm $X \sim N(\mu_X, \sigma_1^2)$ và lấy một mẫu ngẫu nhiên $y_1, y_2,..., y_n$ với cùng một phân phối chuẩn $Y|X = x \sim N(x, \sigma_2^2)$. Khi đó hậu nghiệm $X|Y=y \sim N(\mu, \sigma^2)$ được cho bởi
$$\frac{1}{\sigma^2} = \frac{1}{\sigma_1^2}+\frac{n}{\sigma_2^2}, \frac{\mu}{\sigma^2} = \frac{\mu_X}{\sigma_1^2} + \frac{\sum_{i=1}^n y_i}{\sigma_2^2}.$$

b) Vẽ đồ thị cho phân phối tiên nghiệm và hậu nghiệm trên cùng một hệ trục tọa độ.
