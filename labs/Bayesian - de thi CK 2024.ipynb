1) Tìm hiểu về việc thiết kế một **Naive Bayes Classifier** dựa trên thư viện **sklearn**.

Gaussian Naive Bayes là một kỹ thuật phân loại thống kê dựa trên định lý Bayes, là một trong những thuật toán supervised learning đơn giản nhất. Việc phân loại Naive Bayes có độ chính xác và tốc độ cao trên các tập dữ liệu lớn và nó giả định rằng tác động của một feature trong một lớp là độc lập với các tính năng khác (việc làm này giúp đơn giản hóa việc tính toán và đó là lý do tại sao nó được coi là ngây thơ).

Hãy chạy các code bên dưới để hiểu cách vận hành, từ đó thực hiện yêu cầu ở cuối bài.

```python
# Tự tạo ra một dataset ngẫu nhiên có 800 mẫu với 6 đặc trưng, 3 lớp dựa vào hỗ trợ của thư viện
from sklearn.datasets import make_classification

X, y = make_classification(
    n_features=6,
    n_classes=3,
    n_samples=800,
    n_informative=2,
    random_state=1,
    n_clusters_per_class=1,
)
```

```python
# Vẽ đồ thị scatter plot cho bộ dữ liệu trên
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, marker="*");
```

```python
# Xây dựng tập train và tập test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state = 125
)
```

```python
# Xây dựng một Gaussian NB và huấn luyện nó trên tập train rồi đưa dự liệu vào để dự đoán
from sklearn.naive_bayes import GaussianNB
# Sử dụng model
model = GaussianNB()
# Train model
model.fit(X_train, y_train)
# Danh sách các tiêu chí đánh giá
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)
# Liệt kê danh sách các y dự đoán được
y_pred = model.predict(X_test)
y_pred
```

```python
# Danh sách các y thực tế
y_test
```

```python
# Đánh giá mô hình bằng accuracy và F1-score
accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
# Tiếp tục đánh giá confusion matrix
labels = [0,1,2]
cm = confusion_matrix(y_test, y_pred, labels=labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
```

Bên dưới là sơ đồ mô tả mối liên hệ giữa các biến cố trong hệ thống báo động của một tòa nhà: 
- Bulglary: có kẻ trộm; EarthQK: có động đất.
- Alarm: chuông báo động kêu.
- JohnCall: gọi cho nhân viên John; MaryCall: gọi cho nhân viên Mary.

Các xác suất của biến cố xảy ra được cho trước trong bảng (T là True).

Anh/chị chạy từng cell bên dưới để hiểu về cách xây dựng một network dạng Bayes:

```python
# Dựa vào các mũi tên của sơ đồ, ta định nghĩa được các mối liên hệ trong network như sau:
alarm_model = BayesianNetwork(
    [
        ("Burglary", "Alarm"),
        ("Earthquake", "Alarm"),
        ("Alarm", "JohnCall"),
        ("Alarm", "MaryCall"),
    ]
)

# Xác định các xác suất trong các bảng TabularCPD (conditional probability dependency)
# variable_card viết tắt của từ variable_cardinary, nghĩa là số trạng thái của biến cố

cpd_burglary = TabularCPD(
    variable="Burglary", variable_card=2, values=[[0.999], [0.001]]
)

cpd_earthquake = TabularCPD(
    variable="Earthquake", variable_card=2, values=[[0.998], [0.002]]
)

cpd_alarm = TabularCPD(
    variable="Alarm",
    variable_card=2,
    values=[[0.999, 0.71, 0.06, 0.05], [0.001, 0.29, 0.94, 0.95]],
    evidence=["Burglary", "Earthquake"],
    evidence_card=[2, 2],
)

cpd_johncall = TabularCPD(
    variable="JohnCall",
    variable_card=2,
    values=[[0.95, 0.1], [0.05, 0.9]],
    evidence=["Alarm"],
    evidence_card=[2],
)

cpd_marycall = TabularCPD(
    variable="MaryCall",
    variable_card=2,
    values=[[0.99, 0.3], [0.01, 0.7]],
    evidence=["Alarm"],
    evidence_card=[2],
)
```

```python
# Đưa tất cả các bảng xác suất vào model
alarm_model.add_cpds(
    cpd_burglary, cpd_earthquake, cpd_alarm, cpd_johncall, cpd_marycall
)
```

```python
# Xem các đỉnh có trong network
alarm_model.nodes()
```

```python
# Xem các cạnh có trong network
alarm_model.edges()
```

```python
# In ra xác suất liên quan đến biến cố Alarm
print(cpd_alarm)

# In ra xác suất liên quan đến biến cố MaryCall
print(cpd_marycall)
```

```python
# In tất cả các liên hệ giữa các đỉnh trong network
alarm_model.get_independencies()
```

```python
# In ra các liên hệ giữa các đỉnh khác với đỉnh JohnCall
alarm_model.local_independencies("JohnCall")
```
