$$\frac{1}{\sigma^2} = \frac{1}{\sigma_1^2}+\frac{1}{\sigma_2^2}, \, \frac{\mu}{\sigma^2} = \frac{\mu_X}{\sigma_1^2} + \frac{y}{\sigma_2^2}.$$

$$\frac{1}{\sigma^2} = \frac{1}{\sigma_1^2}+\frac{N}{\sigma_2^2}, \frac{\mu}{\sigma^2} = \frac{\mu_X}{\sigma_1^2} + \frac{\sum_{i=1}^n y_i}{\sigma_2^2}.$$

**Bài 5.** Cho tiên nghiệm $X \sim N(0;1).$ Giả sử rằng $Y|X = x \sim N(x,1).$ Chứng minh rằng phân phối của hậu nghiệm $f_{X|Y}(x|y)$ có phân phối $N(y/2, 1/2).$

*Tham khảo cách giải sau đây*

Ta cần chỉ ra rằng $f_{X|Y}(x|y)$ là phân phối chuẩn với kỳ vọng là $\frac{y}{2}$ và phương sai là $\frac{1}{2}$. Khi đó cần có
\begin{align*}
f_{X|Y}(x|y)=c(y) \exp \left\{-\left(x-\frac{y}{2}\right)^2\right\},
\end{align*}
trong đó $c(y)$ là hàm số theo $y$. Với giá trị $y$ cho trước thì $c(y)$ là hằng số chuẩn hóa để tích phân của $f_{X|Y}(x|y)$ bằng $1$. Theo giả thiết thì
\begin{align*}
f_{Y|X}(y|x)=\frac{1}{\sqrt{2 \pi} } \exp \left\{-\frac{(y-x)^2}{2} \right\},
\end{align*}
\begin{align*}
f_{X}(x)=\frac{1}{\sqrt{2 \pi} } \exp \left\{-\frac{x^2}{2} \right\}.
\end{align*}
Vì thế nên
\begin{align*}
f_{X|Y}(x|y)&=\frac{f_{Y|X}(y|x)f_{X}(x)}{f_{Y}(y)}\\
&= (\textrm{là hàm số theo $y$}) \cdot f_{Y|X}(y|x)f_{X}(x)\\
&=(\textrm{là hàm số theo $y$}) \cdot \exp \left\{-\frac{(y-x)^2+x^2}{2}\right\}\\
&=(\textrm{là hàm số theo $y$}) \cdot \exp \left\{-\left(x-\frac{y}{2}\right)^2+\frac{y^2}{4}\right\}\\
&=(\textrm{là hàm số theo $y$}) \cdot \exp \left\{-\left(x-\frac{y}{2}\right)^2\right\}.
\end{align*}

> **Khoảng tin cậy.** Cho quan sát $Y=y$, khoảng $[a;b]$ được gọi là khoảng tin cậy $(1-\alpha)100 \%$ của $X$ nếu như xác suất để hậu nghiệm của $X$ thuộc vào $[a;b]$ bằng với $1 - \alpha.$ Cụ thể là $$P(a \le X \le b|Y=y) = 1-\alpha.$$
